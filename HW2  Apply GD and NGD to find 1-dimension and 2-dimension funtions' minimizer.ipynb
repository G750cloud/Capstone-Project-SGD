```python
import numpy as np
import pandas as pd

'''
Plain Gradient Decent g(x) = x^2

'''
def grad(x):
    return 2*x
'''
Plain Gradient Decent g(x) = x_1^2 + x_2^2

'''
def grad_2(x_1,x_2):
    return [2*x_1,2*x_2]

'''
Plain Gradient Decent g(x) = (1-x_1)^2 + 100*(x_2-x_1^2)^2

'''
def grad_3(x_1,x_2):
    x_1_grad = 400*x_1*(x_1**2)-400*x_1*x_2+2*x_1-2
    x_2_grad = 200*x_2-200*(x_1**2)
    return [x_1_grad,x_2_grad]

'''
Stochastic Gradient Decent g(x) = x^2-2*x+2

'''
def grad_4(x):
    return 2*x-2

'''
Stochastic Gradient Decent g(x) = 1/2x^2-2*x+1

'''
def grad_5(x):
    return x-2


def plain_gd(grad, cur_x = 0.1, learning_rate = 0.01, precision = 0.0001, max_iters = 10000):
    for i in range(max_iters):
        grad_cur = grad(cur_x)
        if abs(grad_cur) < precision:
            break
        cur_x = cur_x - grad_cur * learning_rate
        print("第",i,"这次迭代：x值为", cur_x)
    print("局部最小值x是：", cur_x)
    return cur_x

def plain_gd_2(grad, cur_x_1 = 0.1, cur_x_2 = 0.2, learning_rate = 0.01, precision = 0.0001, max_iters = 10000):
    for i in range(max_iters):
        grad_cur = grad(cur_x_1,cur_x_2)
        if np.sqrt(grad_cur[0]**2+grad_cur[1]**2) < precision:
            break
        cur_x_1 = cur_x_1 - grad_cur[0] * learning_rate
        cur_x_2 = cur_x_2 - grad_cur[1] * learning_rate
        func_result = (1-cur_x_1)**2+100*(cur_x_2-cur_x_1**2)**2
        print("第",i,"这次迭代：x值为", (cur_x_1,cur_x_2),"结果为",func_result)
    print("局部最小值x是：", (cur_x_1,cur_x_2))
    return cur_x_1,cur_x_2

#plain_gd(grad, cur_x = 8, learning_rate = 0.05, precision = 0.0001, max_iters = 10000)
#plain_gd_2(grad_2, cur_x_1 = 5, cur_x_2 = 7, learning_rate = 0.01, precision = 0.0001, max_iters = 10000)#
#plain_gd_2(grad_3, cur_x_1 = 1, cur_x_2 = 2, learning_rate = 0.001, precision = 0.0001, max_iters = 10000)
#plain_gd(grad_4, cur_x = 8, learning_rate = 0.0005, precision = 0.0001, max_iters = 10000)
plain_gd(grad_5, cur_x = 7, learning_rate = 0.0005, precision = 0.0001, max_iters = 10000)



    
```

```python
import numpy as np
import pandas as pd

'''
Nesterov's accelarated Gradient Decent g(x) = x^2

'''

def N_gd(grad, cur_x1 = 0.2, cur_x2 = 0.1, learning_rate = 0.01, precision = 0.0001, max_iters = 10000):
    for i in range(max_iters):
        y = cur_x2+(i-2)/(i+1)*(cur_x2-cur_x1)      
        grad_cur = grad(y)
        if abs(grad_cur) < precision:
            break
        cur_x1 = cur_x2 
        cur_x2 = y - grad_cur * learning_rate
        print("第",i,"这次迭代：x值为", cur_x2)
    print("局部最小值x是：", cur_x2)
    return cur_x2

def N_gd_2(grad, cur_x1 = [0.2,0.2], cur_x2 = [0.1,0.1], learning_rate = 0.01, precision = 0.0001, max_iters = 10000):
    cur_x1 = np.array(cur_x1)
    cur_x2 = np.array(cur_x2)
    for i in range(max_iters):
        y = cur_x2+(i-2)/(i+1)*(cur_x2-cur_x1)      
        grad_cur = np.array(grad(y[0],y[1]))
        if np.sqrt(grad_cur[0]**2+grad_cur[1]**2) < precision:
            break
        cur_x1 = cur_x2 
        cur_x2 = y - grad_cur * learning_rate
        print("第",i,"这次迭代：x值为", cur_x2)
    print("局部最小值x是：", cur_x2)
    return cur_x2


#N_gd(grad, cur_x1 = 10, cur_x2 = 9.999, learning_rate = 0.0005, precision = 0.0001, max_iters = 10000)#
#N_gd_2(grad_2, cur_x1 = [10,10], cur_x2 = [9,9], learning_rate = 0.01, precision = 0.0001, max_iters = 10000)#
N_gd_2(grad_3, cur_x1 = [1,2], cur_x2 = [1,2.5], learning_rate = 0.001, precision = 0.0001, max_iters = 10000)

#N_gd(grad_4, cur_x1 = 10, cur_x2 = 9.999, learning_rate = 0.0005, precision = 0.0001, max_iters = 10000)#
#N_gd(grad_5, cur_x1 = 10, cur_x2 = 9.999, learning_rate = 0.0005, precision = 0.0001, max_iters = 10000)#
```
